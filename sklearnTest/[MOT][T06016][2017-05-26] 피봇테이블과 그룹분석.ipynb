#### 1.qcut 명령으로 세 개의 나이 그룹을 만든다.
##### //주석 : qcut이 아닌 cut 명령을 사용


```python
titanic = sns.load_dataset("titanic")
titanic
```

```python
bins = [1, 20, 60, 100]
labels = ["미성년", "성년", "노년"]
titanic_age = pd.cut(titanic["age"], bins, labels = labels)
titanic["age_class"] = titanic_age
titanic
```

#### 2.성별, 선실, 나이 그룹에 의한 생존율을 데이터프레임으로 계산한다. 행에는 성별 및 나이 그룹에 대한 다중 인덱스를 사용하고 열에는 선실 인덱스를 사용한다.

```python
df1 = titanic.groupby(['sex', 'age_class', 'class'])["survived"].mean()
df2 = df1.unstack("class")
df2
```

#### 3.성별 및 선실에 의한 생존율을 피봇 데이터 형태로 만든다.

```python
titanic.pivot_table(['survived'], index=['sex', 'class'], aggfunc='mean')
```
